```python
import json, time
from requests import request

data = []
for i in range(5000):
    value_data = '''{
        getRecipeById(id: '''+str(i)+''', imageSizeFilter: \"w480\") {
            id
            tip
            title
            rating
            text
            _votedRating
            recipes_difficult
            typesItem {
                id
                title
            }
            slozhnostItem {
                id
                title
            }
            bludoItem {
                id
                title
            }
            bobrabotkaItem {
                id
                title
            }
            naznachenieItem {
                id
                title
            }
            cooking_time
            rating
            serving
            attachmentsItem {
                cacheItem {
                    filePath
                }
            }
            recipeIngredients {
                id
                title
                value
            }
            recipeInstructions {
                id
                title
                fileItem {
                    cacheItem {
                        filePath
                    }
                }
            }
        }
    }'''
    resp = request('post',f'https://new-api.delikateska.ru/graphql', json = {'query':value_data})
    if resp.status_code == 200:
        if resp.json().get('errors') is None:
            print(resp.json()['data']['getRecipeById']['id'], resp.json()['data']['getRecipeById']['title'])
            data.append(resp.json())
    elif resp.status_code == 500:
        time.sleep(30)
        i = i - 1

with open('data.json', 'w') as outfile:
    json.dump(data, outfile)
```
